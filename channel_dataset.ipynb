```python
#List what datasets are available in the cloud
from intake import open_catalog
url = "https://raw.githubusercontent.com/pangeo-data/pangeo-datastore/master/intake-catalogs/ocean/channel.yaml"
cat = open_catalog(url)
list(cat)
```

Two of the datasets in this list are ours, 'channel_ridge_05km_float_run' which is the model output for this run, and 'channel_ridge_05km_floats' which is the output from the synthetic floats

### First, let's plot the model output (which is on the model grid)

```python
# open the model output and look at the metadata
ds = cat.channel_ridge_05km_float_run.to_dask()
ds
```

```python
#Define grid
from xgcm import Grid 
grid = Grid(ds, coords={'X':{'center':'XC', 'left':'XG'},'Y':{'center':'YC', 'left':'YG'}})
```

```python
#Plot temperature in the surface layer on day 1
ds.THETA.sel(time='0000-01-01').isel(Z=1).plot()
```

This is the model domain. It is an idealized channel with zonal flow over a bump. 

```python
#Find u^2 and v^2 at the first time slice
usq_snap=(ds.UVELSQ).sel(time='0000-01-01')
vsq_snap=(ds.VVELSQ).sel(time='0000-01-01')

#Interpolate onto tracer points and mask where the topography (hFacC) is
usq_snapT=grid.interp(usq_snap,'X').where(ds.hFacC>0)
vsq_snapT=grid.interp(vsq_snap,'Y').where(ds.hFacC>0)

#Plot (u^2+v^2)/2 as a function of depth, mean in y-direction
((usq_snapT+vsq_snapT)/2).mean('YC').plot() 

#The total KE is largest at the surface downstream of the ridge
```

### Now, let's look at the synthetic Argo output

```python
# open the synthetic float output and look at metadata
ds_flt = cat.channel_ridge_05km_floats.to_dask()
ds_flt
```

Take a minute to look at this dataset: it is structured much like Argo data. The depths are given based on the vertical grid points in the model. Each profile has associated temperature, float number (npart), and <time, x, and y> for the downward and upward portions of the trajectory.

The model does not have salinity. 

```python
#Notice the very long dimension profile_index. 
#We cannot plot all the profiles at the same time because the data is too big. 
#instead we will take a couple of profiles and plot them.
ds_flt.isel(profile_index=0).Temperature.plot(y='depth')
```

```python
#plot float positions for first 10 floats
import matplotlib.pyplot as plt
ds_flt.isel(profile_index=slice(0,10)).plot.scatter(x='x_down',y='y_down',c='b')# plot initial positions in blue
ds_flt.isel(profile_index=slice(0,10)).plot.scatter(x='x_up',y='y_up',c='r')# plot final positions in red
plt.xlim(0,2*10**6)
plt.ylim(0,2*10**6)
```

```python
#now plot profiles from first 10 days (this is a heavy operation, but the end result is nice)
import cftime

#define time slice
startslice = cftime.Datetime360Day(year=0, month=1, day=1)
endslice = cftime.Datetime360Day(year=0, month=1, day=10)
timeslice=ds_flt.time_down.where((ds_flt.time_down.values>startslice) & (ds_flt.time_down.values<endslice))

#Plot surface temperature for profiles in timeslice
ds_flt.where(ds_flt.time_down==timeslice,drop=True).isel(depth=1).plot.scatter(x='x_down',y='y_down',hue='Temperature')
```
